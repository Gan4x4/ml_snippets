Рассмотрим самый простой вариант классификатора на два класса — логистическую регрессию:
$$t = b + Wx$$
$$p = \sigma(t) = \frac{1}{1+e^{-t}} = \frac{1}{1+e^{-(b + Wx)}} $$

Выход модели можно интерпретировать как $logit$  - функцию от вероятности 

$$logit(p) = \ln \frac {p} {1 - p} = b + Wx$$

Которая лучше позволяет преедполагать аддитивность вкладов признаков.
 
Для многоклассовой [логистической регрессии](https://en.wikipedia.org/wiki/Logistic_regression):

$$logit(p_i) = \ln \frac {p_i} {1 - p_i}$$

Как выразить вероятность через logit ?

Напишите функцию, которая переводит логиты в вероятности. Формат вывода `dict{'NEGATIVE': val, 'POSITIVE': val}` с точностью значений 5 знаков после запятой.

Вариант №1 



````python
from scipy.special import softmax

def logit_to_prob(logits): 
  prob = softmax(logits)
  return {'NEGATIVE': round(prob[0],5), 'POSITIVE': round(prob[1],5)} 
````

Вариант №2

Выразим вероятность через логиты:

$$logit(p) = y =  \ln \frac {p} {1 - p}$$

$$ e^y = e^{\ln \frac {p} {1 - p}}$$

$$ e^y = \frac {p} {1 - p}$$

$$ e^y-e^yp = p$$

$$ e^y = p + e^yp $$

$$ e^y = p(1 + e^y) $$

$$ p = \frac {e^y} {1 + e^y} $$



````python
import numpy as np
def logit_to_prob(logits): 
  odds = np.exp(logits)
  prob = odds/(1+odds)
  return {'NEGATIVE': round(prob[0],5), 'POSITIVE': round(prob[1],5)} 
````
